```python
import pandas as pd
import os

# Ensure the output directory exists
output_dir = "C:/Users/19788/Downloads"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Read CSV with specified data types
file_path = "combined_output.csv"
df = pd.read_csv(file_path, dtype=str, low_memory=False)

# Convert "Incident Date" to datetime format and extract the year
df["Incident Date"] = pd.to_datetime(df["Incident Date"], errors='coerce')
df["Year"] = df["Incident Date"].dt.year  # Extract year

# Perform the analysis
selected_columns = ["Year", "Incident Date", "Incident Time", "Flight ID", "Laser Color", "Aircraft", "Altitude", "Airport", "Injury", "City", "State"]
df = df[selected_columns]  # Keep only necessary columns

color_counts = df.groupby(["Year", "Laser Color"]).size().reset_index(name="Count")

# Get top 4 most frequent laser colors per year while keeping the selected features
top_colors_per_year = color_counts.groupby("Year", group_keys=False).apply(lambda x: x.nlargest(4, "Count"))
top_colors_per_year = top_colors_per_year.merge(df, on=["Year", "Laser Color"], how="left").drop_duplicates()

# Save results to Excel
output_path = os.path.join(output_dir, "top_laser_colors.xlsx")
with pd.ExcelWriter(output_path, engine="xlsxwriter") as writer:
    for year, data in top_colors_per_year.groupby("Year"):
        data.drop(columns=["Year"], inplace=True)  # Drop year column for cleaner output
        data.to_excel(writer, sheet_name=str(year), index=False)

print(f"Excel file saved at: {output_path}")
```
